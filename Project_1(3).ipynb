```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load grayscale image
img = cv2.imread('Image 1 Denoised.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate histogram
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Display the images side by side
plt.figure(figsize=(10, 5))

# Plot grayscale image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Image 1 Denoised')

# Plot histogram using plt.bar() with blue color
plt.subplot(1, 2, 2)
pixel_values = np.arange(256)
plt.bar(pixel_values, hist.ravel())  # Change color to blue
plt.title('Histogram of Denoised Image 1')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
