```python
import cv2
import matplotlib.pyplot as plt

def convert_to_grayscale(image):
    # Convert the image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return grayscale_image

def resize_with_aspect_ratio(image, width_percent):
    # Get the original image dimensions
    original_height, original_width = image.shape[:2]

    # Calculate the new width
    new_width = int(original_width * (width_percent / 100))

    # Calculate the aspect ratio
    aspect_ratio = new_width / original_width

    # Calculate the new height using the aspect ratio
    new_height = int(original_height * aspect_ratio)

    # Resize the image
    resized_image = cv2.resize(image, (new_width, new_height))
    return resized_image

# Load the original image
original_image_path = '1.jpg'  # Replace 'your_image.jpg' with the path to your image
original_image = cv2.imread(original_image_path)

# Convert the original image to grayscale
gray_original_image = convert_to_grayscale(original_image)

# Resize the grayscale image with aspect ratio (e.g., resizing to 50%)
resized_image = resize_with_aspect_ratio(gray_original_image, width_percent=50)

# Display the images side by side
plt.figure(figsize=(10, 5))

# Display the original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(gray_original_image, cmap='gray')
plt.title("Original Grayscale Image")

# Display the resized grayscale image
plt.subplot(1, 2, 2)
plt.imshow(resized_image, cmap='gray')
plt.title("Resized Grayscale Image")

# Save the resized image
resized_image_path = 'Image 1 Resized.jpg'
cv2.imwrite(resized_image_path, resized_image)
print("Resized image saved successfully.")
plt.show()
```
