```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('Image 1 Resized.jpg')

# Apply non-local means denoising to the resized image
denoised_image = cv2.fastNlMeansDenoising(img, None, h=10, templateWindowSize=7, searchWindowSize=21)

# Display the images side by side
plt.figure(figsize=(10, 5))

# Display the resized grayscale image
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title("Resized Image 1 (Before Denoising)")

# Display the denoised image
plt.subplot(1, 2, 2)
plt.imshow(denoised_image)
plt.title("Resized Image 1 (After Denoising)")

# Save the denoised image
denoised_image_path = 'Image 1 Denoised.jpg'
cv2.imwrite(denoised_image_path, denoised_image)
print("Denoised image saved successfully.")


plt.show()
```
